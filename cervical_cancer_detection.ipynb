```python
target_df=cancer_df['Biopsy']
input_df=cancer_df.drop(columns=['Biopsy'])
target_df.shape
```

```python
input_df.shape
```

```python
X=np.array(input_df).astype('float32')
y=np.array(target_df).astype('float32')
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_test,X_val,y_test,y_val=train_test_split(X_test,y_test,test_size=0.5)
```

```python
import xgboost as xgb
model=xgb.XGBClassifier(learning_rate=0.1,max_depth=50,n_estimators=100)
model.fit(X_train,y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_estimators':[50,100,200],
    'max_depth':[3,5,7,10],
    'learning_rate':[0.01,0.05,0.1,0.2]
}

random_search = RandomizedSearchCV(
    estimator=xgb.XGBClassifier(),
    param_distributions=param_grid,
    n_iter=10,
    cv=5,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train,y_train)

best_model = random_search.best_estimator_

print(random_search.best_params_)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred_best = best_model.predict(X_test)

print(
    classification_report(
        y_test,
        y_pred_best
    )
)

print(
    "Accuracy:",
    best_model.score(X_test,y_test)
)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    best_model,
    X,
    y,
    cv=5,
    scoring='accuracy'
)

print("Cross Validation Scores:")
print(scores)

print("Mean Accuracy:", scores.mean())
```

```python
result_train=model.score(X_train,y_train)
result_train
```

```python
result_test=model.score(X_test,y_test)
result_test
```

```python
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.metrics import roc_auc_score

y_prob = best_model.predict_proba(X_test)[:,1]

roc_auc = roc_auc_score(y_test,y_prob)

print("ROC-AUC Score:", roc_auc)
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(
    y_test,
    y_prob
)

plt.figure(figsize=(8,6))

plt.plot(
    fpr,
    tpr,
    label=f"AUC = {roc_auc:.3f}"
)

plt.plot([0,1],[0,1],'--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")

plt.legend()

plt.show()
```

```python
cm=confusion_matrix(y_predict,y_test)
sns.heatmap(cm,annot=True)
```

```python
importance = pd.Series(
    best_model.feature_importances_,
    index=input_df.columns
)

top_features = importance.sort_values(
    ascending=False
).head(10)

plt.figure(figsize=(10,6))

top_features.plot(kind='bar')

plt.title("Top 10 Important Features")

plt.ylabel("Importance Score")

plt.show()
```

```python
import pandas as pd

print("Enter Patient Details:\n")

age = float(input("Age: "))
partners = float(input("Number of sexual partners: "))
first_sex = float(input("First sexual intercourse age: "))
pregnancies = float(input("Number of pregnancies: "))
smokes = float(input("Smokes (0/1): "))
smokes_years = float(input("Smokes (years): "))
smokes_packs = float(input("Smokes (packs/year): "))
hormonal = float(input("Hormonal Contraceptives (0/1): "))
hormonal_years = float(input("Hormonal Contraceptives (years): "))
iud = float(input("IUD (0/1): "))
iud_years = float(input("IUD (years): "))
stds = float(input("STDs (0/1): "))
stds_num = float(input("STDs (number): "))
stds_condy = float(input("STDs:condylomatosis (0/1): "))
stds_cervical = float(input("STDs:cervical condylomatosis (0/1): "))
stds_vaginal = float(input("STDs:vaginal condylomatosis (0/1): "))
stds_vulvo = float(input("STDs:vulvo-perineal condylomatosis (0/1): "))
stds_syphilis = float(input("STDs:syphilis (0/1): "))
stds_pid = float(input("STDs:pelvic inflammatory disease (0/1): "))
stds_herpes = float(input("STDs:genital herpes (0/1): "))
stds_molluscum = float(input("STDs:molluscum contagiosum (0/1): "))
stds_aids = float(input("STDs:AIDS (0/1): "))
stds_hiv = float(input("STDs:HIV (0/1): "))
stds_hepb = float(input("STDs:Hepatitis B (0/1): "))
stds_hpv = float(input("STDs:HPV (0/1): "))
stds_diag = float(input("STDs: Number of diagnosis: "))
dx_cancer = float(input("Dx:Cancer (0/1): "))
dx_cin = float(input("Dx:CIN (0/1): "))
dx_hpv = float(input("Dx:HPV (0/1): "))
dx = float(input("Dx (0/1): "))
hinselmann = float(input("Hinselmann (0/1): "))
schiller = float(input("Schiller (0/1): "))
citology = float(input("Citology (0/1): "))

patient = pd.DataFrame([[

    age,
    partners,
    first_sex,
    pregnancies,
    smokes,
    smokes_years,
    smokes_packs,
    hormonal,
    hormonal_years,
    iud,
    iud_years,
    stds,
    stds_num,
    stds_condy,
    stds_cervical,
    stds_vaginal,
    stds_vulvo,
    stds_syphilis,
    stds_pid,
    stds_herpes,
    stds_molluscum,
    stds_aids,
    stds_hiv,
    stds_hepb,
    stds_hpv,
    stds_diag,
    dx_cancer,
    dx_cin,
    dx_hpv,
    dx,
    hinselmann,
    schiller,
    citology

]], columns=input_df.columns)

patient_scaled = scaler.transform(patient)

prediction = best_model.predict(patient_scaled)
probability = best_model.predict_proba(patient_scaled)[0][1]

print("\nRESULT")
print(f"Cancer Risk Probability: {probability*100:.2f}%")

if prediction[0] == 1:
    print("⚠️ High Risk of Cervical Cancer")
else:
    print("✅ Low Risk of Cervical Cancer")
```
